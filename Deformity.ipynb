```python
import numpy as np
import pandas as pd
import sympy as sym
import math
import matplotlib.pyplot as plt
import mpmath

from sympy.interactive import printing
printing.init_printing(use_latex=True)
```

```python
x=np.arange(-5, 5, 0.1)
f=[(lambda y: -1*y**3+3*y**2+10*y-24)(y) for y in x]


fig, ax=plt.subplots(1,1,figsize=(18,18))
ax.plot(x, f)
ax.axhline(0, color='#FF0000')

v=mpmath.polyroots([-1,3,10,-24])
for i, x in enumerate(v):
    print(x)
    ax.axvline(x, color='#000000')
    
plt.show()
```

```python
s=-math.pi
f=[]
x=[]
while (s<math.pi+0.001):
    f.append(np.sin(s))
    x.append(s)
    s+=0.02*math.pi

fig, ax=plt.subplots(3, 1, figsize=(18,18))

ax[0].set_title('Krok co 0.02*pi')
ax[0].plot(x, f)

ax[1].set_title('Krok co 0.2*pi')
ax[1].plot(x[::10], f[::10])

ax[2].set_title('Krok co pi')
ax[2].plot(x[::50], f[::50])
ax[2].set_ylim(-1, 1)


plt.show()
```
